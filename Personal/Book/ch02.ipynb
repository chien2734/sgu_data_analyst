#### Duck typing

```python
def isiterable(obj):
  try:
    iter(obj)
    return True
  except TypeError: # not iterable
    return False
```

```python
isiterable("a string") #True
isiterable([1, 2, 3]) #True
isiterable(5) #False
```

#### for loops

```python
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
  if value is None:
    continue
  total += value
print(total)
```

```python
sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0
for value in sequence:
  if value == 5:
    break
  total_until_5 += value
print(total_until_5)
```

```python
for i in range(4):
   for j in range(4):
    if j > i:
      break
    print((i, j))
```

#### while loops

```python
x = 256
total = 0
while x > 0:
  if total > 500:
    break
  total += x
  x = x // 2
print(total)
```
