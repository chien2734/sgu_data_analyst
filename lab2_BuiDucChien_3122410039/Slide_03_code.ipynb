### Reshaping Data

```python
import pandas as pd
data = {
    "Tên": ["An", "An", "Bình", "Bình"],
    "Môn": ["Toán", "Văn", "Toán", "Văn"],
    "Điểm": [8, 7, 9, 6]
}

df = pd.DataFrame(data)
#print(df)

#pivot : long->wide
df_pivot = df.pivot(index="Tên", columns="Môn", values="Điểm")
#print(df_pivot)

#Melt : wide->long
df_melt = df_pivot.reset_index().melt(id_vars="Tên", value_vars=["Toán", "Văn"], var_name="Môn", value_name="Điểm")
#print(df_melt)

# Stack & Unstack
stacked = df_pivot.stack()
#print(stacked)
unstacked = stacked.unstack()
#print(unstacked)

#Concat
df1 = pd.DataFrame({"A": [1, 2]})
df2 = pd.DataFrame({"A": [3, 4]})
df_concat = pd.concat([df1, df2], ignore_index=True)
#print(df_concat)

# Merge
df_left = pd.DataFrame({"id": [1, 2], "Tên": ["An", "Bình"]})
df_right = pd.DataFrame({"id": [1, 2], "Tuổi": [20, 21]})
df_merge = pd.merge(df_left, df_right, on="id")
print(df_merge)
```

### Subset Observations(Rows)

```python
import pandas as pd
data = {
    "Tên": ["An", "An", "Bình", "Bình"],
    "Môn": ["Toán", "Văn", "Toán", "Văn"],
    "Điểm": [8, 7, 9, 6]
}
df = pd.DataFrame(data)
#print(df)

#iloc
print(df.iloc[0])
print(df.iloc[[0, 2]])

#loc
df2 = df.set_index("Tên")
print(df2.loc["An"])

#Select row by condition
print(df[df["Điểm"] > 7]) # Lấy những người có điểm > 8

#Query
print(df.query("Điểm > 7 and Điểm <= 9")) #Lấy những người có điểm > 7 và <=9
# Select first n rows
df.head(2)
#Select last n rows
df.tail(1)
#Randomly select fraction of rows
df.sample(frac=0.25)
#Randomly select n rows
df.sample(n=2)
```

### Subset Variables (Columns)

```python
#select multiple columns with specific names
df[["Tên", "Điểm"]]

#select single column with specific name
df["Tên"]

#select columns whose name matches regular expression regex
df.filter(like= "Điểm")
```

### Method chaining


```python
# filter score > 5, add new column, sort by score descending
result = (
    df.query("Điểm > 5")
      .assign(Xếp_loại=lambda x: ["Giỏi" if d >= 8 else "Khá" for d in x["Điểm"]])
      .sort_values("Điểm", ascending=False)
)

print(result)
```

## Python Finance Application with Pandas

```python
import pandas as pd
# Create a DataFrame to store transactions
columns = ['Date', 'Description', 'Amount']
ledger = pd.DataFrame(columns=columns)
# Function to add a transaction
def add_transaction(date, description, amount):
  global ledger
  new_transaction = pd.DataFrame([[date, description, amount]], columns=columns)
  ledger = pd.concat([ledger, new_transaction], ignore_index=True)
# Function to view the ledger
def view_ledger():
  print(ledger)
# Function to get the current balance
def get_balance():
  return ledger['Amount'].sum()
# Adding sample transactions
add_transaction('2023-11-01', 'Income', 1000)
add_transaction('2023-11-02', 'Groceries', -200)
add_transaction('2023-11-03', 'Utilities', -100)
# Viewing the ledger
view_ledger()
# Checking the current balance
print("Current Balance:", get_balance())
```
