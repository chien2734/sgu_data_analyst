**1. Ph√¢n t√≠ch ƒë∆°n bi·∫øn (Univariate Analysis) l√† g√¨? Kh√°c g√¨ v·ªõi ph√¢n t√≠ch hai bi·∫øn (Bivariate Analysis)?**
* Ph√¢n t√≠ch ƒë∆°n bi·∫øn (Univariate Analysis):
  * Xem x√©t v√† m√¥ t·∫£ ch·ªâ m·ªôt bi·∫øn duy nh·∫•t trong t·∫≠p d·ªØ li·ªáu.
  * M·ª•c ti√™u: t√≥m t·∫Øt d·ªØ li·ªáu, t√¨m hi·ªÉu ph√¢n ph·ªëi, ph√°t hi·ªán gi√° tr·ªã ngo·∫°i lai.
  * V√≠ d·ª•: ph√¢n t√≠ch ph√¢n ph·ªëi chi·ªÅu cao c·ªßa sinh vi√™n.
* Ph√¢n t√≠ch hai bi·∫øn (Bivariate Analysis):
  * Xem x√©t m·ªëi quan h·ªá gi·ªØa hai bi·∫øn c√πng l√∫c.
  * M·ª•c ti√™u: t√¨m ra m·ªëi li√™n h·ªá, m·ª©c ƒë·ªô t∆∞∆°ng quan ho·∫∑c nh√¢n qu·∫£.
  * V√≠ d·ª•: chi·ªÅu cao c√≥ li√™n quan th·∫ø n√†o ƒë·∫øn c√¢n n·∫∑ng?

**2. C√°c th∆∞·ªõc ƒëo th·ªëng k√™ th∆∞·ªùng d√πng trong ph√¢n t√≠ch ƒë∆°n bi·∫øn**
* Trung t√¢m d·ªØ li·ªáu (Central tendency):
  * Mean (trung b√¨nh)
  * Median (trung v·ªã)
  * Mode (gi√° tr·ªã xu·∫•t hi·ªán nhi·ªÅu nh·∫•t)
* ƒê·ªô ph√¢n t√°n (Dispersion):
  * Range (kho·∫£ng gi√° tr·ªã)
  * Variance (ph∆∞∆°ng sai)
  * Standard Deviation (ƒë·ªô l·ªách chu·∫©n)
  * IQR (Interquartile Range ‚Äì kho·∫£ng t·ª© ph√¢n v·ªã)
* H√¨nh d·∫°ng ph√¢n ph·ªëi:
  * Skewness (ƒë·ªô l·ªách)
  * Kurtosis (ƒë·ªô nh·ªçn)

**3. Trong ph√¢n t√≠ch hai bi·∫øn, l√†m th·∫ø n√†o ƒë·ªÉ x√°c ƒë·ªãnh m·ªëi quan h·ªá gi·ªØa hai bi·∫øn?**
* T∆∞∆°ng quan (Correlation): ƒëo l∆∞·ªùng m·ª©c ƒë·ªô hai bi·∫øn thay ƒë·ªïi c√πng nhau (th∆∞·ªùng d√πng h·ªá s·ªë Pearson, Spearman).
* Bi·ªÉu ƒë·ªì: scatter plot, heatmap, boxplot‚Ä¶
* Th·ªëng k√™ ki·ªÉm ƒë·ªãnh: ki·ªÉm ƒë·ªãnh Chi-square (cho bi·∫øn ph√¢n lo·∫°i), ANOVA (so s√°nh trung b√¨nh nhi·ªÅu nh√≥m).
* Nh√¢n qu·∫£ (Causality): c·∫ßn thi·∫øt k·∫ø th√≠ nghi·ªám ho·∫∑c s·ª≠ d·ª•ng c√°c m√¥ h√¨nh nh√¢n qu·∫£ (v√≠ d·ª•: DAG, Granger causality trong chu·ªói th·ªùi gian).

**4. S·ª± kh√°c bi·ªát gi·ªØa t∆∞∆°ng quan (correlation) v√† hi·ªáp bi·∫øn (covariance)**
* Covariance (Hi·ªáp bi·∫øn):
  * ƒêo l∆∞·ªùng c√°ch hai bi·∫øn thay ƒë·ªïi c√πng nhau.
  * C√≥ th·ªÉ d∆∞∆°ng, √¢m ho·∫∑c b·∫±ng 0.
  * Kh√¥ng chu·∫©n h√≥a ‚Üí kh√≥ so s√°nh gi·ªØa c√°c c·∫∑p bi·∫øn kh√°c nhau.
* Correlation (T∆∞∆°ng quan):
  * Chu·∫©n h√≥a hi·ªáp bi·∫øn trong kho·∫£ng \[-1, 1\].
  * D·ªÖ gi·∫£i th√≠ch:
    * 1: t∆∞∆°ng quan thu·∫≠n tuy·ªát ƒë·ªëi
    * -1: t∆∞∆°ng quan ngh·ªãch tuy·ªát ƒë·ªëi
    * 0: kh√¥ng c√≥ t∆∞∆°ng quan tuy·∫øn t√≠nh

üëâ T√≥m g·ªçn: Correlation l√† d·∫°ng chu·∫©n h√≥a c·ªßa Covariance.

**5. Khi n√†o n√™n d√πng bi·ªÉu ƒë·ªì tr·ª±c quan?**
* Ph√¢n t√≠ch ƒë∆°n bi·∫øn:
  * Histogram (ph√¢n ph·ªëi t·∫ßn su·∫•t)
  * Boxplot (ph√°t hi·ªán ngo·∫°i lai)
  * Density plot
* Ph√¢n t√≠ch hai bi·∫øn:
  * Scatter plot (2 bi·∫øn s·ªë li√™n t·ª•c)
  * Heatmap (t∆∞∆°ng quan nhi·ªÅu bi·∫øn s·ªë)
  * Boxplot/Violin plot (bi·∫øn s·ªë + bi·∫øn ph√¢n lo·∫°i)


```python
# 6. Code m·∫´u scatter plot v√† heatmap trong Python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# V√≠ d·ª• dataset
df = sns.load_dataset("iris")

# Scatter plot: m·ªëi quan h·ªá gi·ªØa chi·ªÅu d√†i v√† chi·ªÅu r·ªông c√°nh hoa
sns.scatterplot(x="sepal_length", y="sepal_width", data=df)
plt.title("Scatter Plot gi·ªØa Sepal Length v√† Sepal Width")
plt.show()

# Heatmap: ma tr·∫≠n t∆∞∆°ng quan
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap - Ma tr·∫≠n t∆∞∆°ng quan")
plt.show()
```

```python
# 7. Tr·ª±c quan h√≥a bi·∫øn s·ªë v√† bi·∫øn ph√¢n lo·∫°i (Boxplot, Violin plot)
# Boxplot: so s√°nh sepal_width theo lo√†i hoa
sns.boxplot(x="species", y="sepal_width", data=df)
plt.title("Boxplot - Sepal width theo Species")
plt.show()

# Violin plot: so s√°nh ph√¢n ph·ªëi sepal_width theo lo√†i hoa
sns.violinplot(x="species", y="sepal_width", data=df)
plt.title("Violin Plot - Sepal width theo Species")
plt.show()
```

### **Nhi·ªám v·ª• 1**: Ph√¢n t√≠ch d·ªØ li·ªáu ƒë∆°n bi·∫øn tr√™n d·ªØ li·ªáu v·ªÅ chim c√°nh c·ª•t

#### 1. Import th∆∞ vi·ªán v√† n·∫°p d·ªØ li·ªáu

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
penguins = pd.read_csv("data/penguins_size.csv")
penguins = penguins[['species', 'culmen_length_mm']]
```

#### 2. Ph√¢n t√≠ch ƒë∆°n bi·∫øn b·∫±ng Histogram

```python
# case 1: basic
sns.histplot( data = penguins, x= penguins["culmen_length_mm"])
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.histplot( data = penguins, x=penguins["culmen_length_mm"])
ax.set_xlabel('Culmen Length in mm',fontsize = 15)
ax.set_ylabel('Count of records', fontsize = 15)
ax.set_title('Univariate analysis of Culmen Length',fontsize= 20)
plt.savefig("Histogram.png")
```

#### 3. Ph√¢n t√≠ch ƒë∆°n bi·∫øn b·∫±ng bar chart

```python
# case 1: basic
sns.countplot(data = penguins, x= penguins['species'])
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.countplot(data = penguins, x= penguins['species'])
ax.set_xlabel('Penguin Species',fontsize = 15)
ax.set_ylabel('Count of records',fontsize = 15)
ax.set_title('Univariate analysis of Penguin Species', fontsize= 20)
ax.set_title('Univariate analysis of Culmen Length', fontsize= 20)
plt.savefig("Bar_chart.png")
```

#### 4. Ph√¢n t√≠ch ƒë∆°n bi·∫øn b·∫±ng bi·ªÉu ƒë·ªì tr√≤n (Pie-chart)

```python
penguins_group = penguins.groupby('species').count()
penguins_group = penguins_group.reset_index()
# case 1: basic
plt.pie(penguins_group["culmen_length_mm"], labels =
penguins_group['species'])
plt.show()
# case 2: advanced
cols = ['g', 'b', 'r']
plt.pie(penguins_group["culmen_length_mm"], labels =
penguins_group['species'],colors = cols)
plt.title('Univariate Analysis of Species', fontsize=15)
plt.show()
plt.savefig("Pie_chart.png")
```

### **Nhi·ªám v·ª• 2**: Ph√¢n t√≠ch d·ªØ li·ªáu ƒë∆°n bi·∫øn tr√™n d·ªØ li·ªáu gi√° nh√†

#### 1. Import th∆∞ vi·ªán, n·∫°p d·ªØ li·ªáu gi√° nh√† v√† ph√¢n t√≠ch ƒë∆°n bi·∫øn d·ª±a v√†o boxplot

```python
houseprices_data = pd.read_csv("data/HousingPrices.csv")
houseprices_data = houseprices_data[['Zip','Price','Area','Room']]
# case 1: Create a boxplot using the boxplot method
sns.boxplot(data = houseprices_data, x= houseprices_data["Price"])
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.boxplot(data = houseprices_data, x= houseprices_data["Price"])
ax.set_xlabel('House Prices in millions',fontsize = 15)
ax.set_title('Univariate analysis of House Prices', fontsize= 20)
plt.ticklabel_format(style='plain', axis='x')
plt.savefig("Boxplot.png")
```
