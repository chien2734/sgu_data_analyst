* Ph√¢n t√≠ch hai bi·∫øn (Bivariate Analysis):

* T∆∞∆°ng quan (Correlation): ƒëo l∆∞·ªùng m·ª©c ƒë·ªô hai bi·∫øn thay ƒë·ªïi c√πng nhau (th∆∞·ªùng d√πng h·ªá s·ªë Pearson, Spearman).

**4. S·ª± kh√°c bi·ªát gi·ªØa t∆∞∆°ng quan (correlation) v√† hi·ªáp bi·∫øn (covariance)**
* Covariance (Hi·ªáp bi·∫øn):
  * ƒêo l∆∞·ªùng c√°ch hai bi·∫øn thay ƒë·ªïi c√πng nhau.
  * C√≥ th·ªÉ d∆∞∆°ng, √¢m ho·∫∑c b·∫±ng 0.
  * Kh√¥ng chu·∫©n h√≥a ‚Üí kh√≥ so s√°nh gi·ªØa c√°c c·∫∑p bi·∫øn kh√°c nhau.
* Correlation (T∆∞∆°ng quan):
  * Chu·∫©n h√≥a hi·ªáp bi·∫øn trong kho·∫£ng \[-1, 1\].
  * D·ªÖ gi·∫£i th√≠ch:
    * 1: t∆∞∆°ng quan thu·∫≠n tuy·ªát ƒë·ªëi
    * -1: t∆∞∆°ng quan ngh·ªãch tuy·ªát ƒë·ªëi
    * 0: kh√¥ng c√≥ t∆∞∆°ng quan tuy·∫øn t√≠nh

üëâ T√≥m g·ªçn: Correlation l√† d·∫°ng chu·∫©n h√≥a c·ªßa Covariance.

**5. Khi n√†o n√™n d√πng bi·ªÉu ƒë·ªì tr·ª±c quan?**
* Ph√¢n t√≠ch ƒë∆°n bi·∫øn:
  * Histogram (ph√¢n ph·ªëi t·∫ßn su·∫•t)
  * Boxplot (ph√°t hi·ªán ngo·∫°i lai)
  * Density plot
* Ph√¢n t√≠ch hai bi·∫øn:
  * Scatter plot (2 bi·∫øn s·ªë li√™n t·ª•c)
  * Heatmap (t∆∞∆°ng quan nhi·ªÅu bi·∫øn s·ªë)
  * Boxplot/Violin plot (bi·∫øn s·ªë + bi·∫øn ph√¢n lo·∫°i)


```python
# 6. Code m·∫´u scatter plot v√† heatmap trong Python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# V√≠ d·ª• dataset
df = sns.load_dataset("iris")

# Scatter plot: m·ªëi quan h·ªá gi·ªØa chi·ªÅu d√†i v√† chi·ªÅu r·ªông c√°nh hoa
sns.scatterplot(x="sepal_length", y="sepal_width", data=df)
plt.title("Scatter Plot gi·ªØa Sepal Length v√† Sepal Width")
plt.show()

# Heatmap: ma tr·∫≠n t∆∞∆°ng quan
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap - Ma tr·∫≠n t∆∞∆°ng quan")
plt.show()
```

```python
# 7. Tr·ª±c quan h√≥a bi·∫øn s·ªë v√† bi·∫øn ph√¢n lo·∫°i (Boxplot, Violin plot)
# Boxplot: so s√°nh sepal_width theo lo√†i hoa
sns.boxplot(x="species", y="sepal_width", data=df)
plt.title("Boxplot - Sepal width theo Species")
plt.show()

# Violin plot: so s√°nh ph√¢n ph·ªëi sepal_width theo lo√†i hoa
sns.violinplot(x="species", y="sepal_width", data=df)
plt.title("Violin Plot - Sepal width theo Species")
plt.show()
```
