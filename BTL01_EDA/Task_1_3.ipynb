##**1.3.1. Ôn lý thuyết**

**1. Phân tích đơn biến (Univariate Analysis) là gì? Khác gì với phân tích hai biến (Bivariate Analysis)?**
* Phân tích đơn biến (Univariate Analysis):
  * Xem xét và mô tả chỉ một biến duy nhất trong tập dữ liệu.
  * Mục tiêu: tóm tắt dữ liệu, tìm hiểu phân phối, phát hiện giá trị ngoại lai.
  * Ví dụ: phân tích phân phối chiều cao của sinh viên.
* Phân tích hai biến (Bivariate Analysis):
  * Xem xét mối quan hệ giữa hai biến cùng lúc.
  * Mục tiêu: tìm ra mối liên hệ, mức độ tương quan hoặc nhân quả.
  * Ví dụ: chiều cao có liên quan thế nào đến cân nặng?

**2. Các thước đo thống kê thường dùng trong phân tích đơn biến**
* Trung tâm dữ liệu (Central tendency):
  * Mean (trung bình)
  * Median (trung vị)
  * Mode (giá trị xuất hiện nhiều nhất)
* Độ phân tán (Dispersion):
  * Range (khoảng giá trị)
  * Variance (phương sai)
  * Standard Deviation (độ lệch chuẩn)
  * IQR (Interquartile Range – khoảng tứ phân vị)
* Hình dạng phân phối:
  * Skewness (độ lệch)
  * Kurtosis (độ nhọn)

**3. Trong phân tích hai biến, làm thế nào để xác định mối quan hệ giữa hai biến?**
* Tương quan (Correlation): đo lường mức độ hai biến thay đổi cùng nhau (thường dùng hệ số Pearson, Spearman).
* Biểu đồ: scatter plot, heatmap, boxplot…
* Thống kê kiểm định: kiểm định Chi-square (cho biến phân loại), ANOVA (so sánh trung bình nhiều nhóm).
* Nhân quả (Causality): cần thiết kế thí nghiệm hoặc sử dụng các mô hình nhân quả (ví dụ: DAG, Granger causality trong chuỗi thời gian).

**4. Sự khác biệt giữa tương quan (correlation) và hiệp biến (covariance)**
* Covariance (Hiệp biến):
  * Đo lường cách hai biến thay đổi cùng nhau.
  * Có thể dương, âm hoặc bằng 0.
  * Không chuẩn hóa → khó so sánh giữa các cặp biến khác nhau.
* Correlation (Tương quan):
  * Chuẩn hóa hiệp biến trong khoảng \[-1, 1\].
  * Dễ giải thích:
    * 1: tương quan thuận tuyệt đối
    * -1: tương quan nghịch tuyệt đối
    * 0: không có tương quan tuyến tính

👉 Tóm gọn: Correlation là dạng chuẩn hóa của Covariance.

**5. Khi nào nên dùng biểu đồ trực quan?**
* Phân tích đơn biến:
  * Histogram (phân phối tần suất)
  * Boxplot (phát hiện ngoại lai)
  * Density plot
* Phân tích hai biến:
  * Scatter plot (2 biến số liên tục)
  * Heatmap (tương quan nhiều biến số)
  * Boxplot/Violin plot (biến số + biến phân loại)


```python
# 6. Code mẫu scatter plot và heatmap trong Python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ví dụ dataset
df = sns.load_dataset("iris")

# Scatter plot: mối quan hệ giữa chiều dài và chiều rộng cánh hoa
sns.scatterplot(x="sepal_length", y="sepal_width", data=df)
plt.title("Scatter Plot giữa Sepal Length và Sepal Width")
plt.show()

# Heatmap: ma trận tương quan
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap - Ma trận tương quan")
plt.show()
```

```python
# 7. Trực quan hóa biến số và biến phân loại (Boxplot, Violin plot)
# Boxplot: so sánh sepal_width theo loài hoa
sns.boxplot(x="species", y="sepal_width", data=df)
plt.title("Boxplot - Sepal width theo Species")
plt.show()

# Violin plot: so sánh phân phối sepal_width theo loài hoa
sns.violinplot(x="species", y="sepal_width", data=df)
plt.title("Violin Plot - Sepal width theo Species")
plt.show()
```

#### **Nhiệm vụ 1**: Phân tích dữ liệu đơn biến trên dữ liệu về chim cánh cụt

##### 1. Import thư viện và nạp dữ liệu

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
penguins = pd.read_csv("penguins_size.csv")
penguins = penguins[['species', 'culmen_length_mm']]
```

##### 2. Phân tích đơn biến bằng Histogram

```python
# case 1: basic
sns.histplot( data = penguins, x= penguins["culmen_length_mm"])
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.histplot( data = penguins, x=penguins["culmen_length_mm"])
ax.set_xlabel('Culmen Length in mm',fontsize = 15)
ax.set_ylabel('Count of records', fontsize = 15)
ax.set_title('Univariate analysis of Culmen Length',fontsize= 20)
plt.savefig("Histogram.png")
```

##### 3. Phân tích đơn biến bằng bar chart

```python
# case 1: basic
sns.countplot(data = penguins, x= penguins['species'])
# case 2: advanced
plt.figure(figsize= (12,6))
ax = sns.countplot(data = penguins, x= penguins['species'])
ax.set_xlabel('Penguin Species',fontsize = 15)
ax.set_ylabel('Count of records',fontsize = 15)
ax.set_title('Univariate analysis of Penguin Species', fontsize= 20)
ax.set_title('Univariate analysis of Culmen Length', fontsize= 20)
plt.savefig("Bar_chart.png")
```

##### 4. Phân tích đơn biến bằng biểu đồ tròn (Pie-chart)

```python
penguins_group = penguins.groupby('species').count()
penguins_group = penguins_group.reset_index()
# case 1: basic
plt.pie(penguins_group["culmen_length_mm"], labels =
penguins_group['species'])
plt.show()
# case 2: advanced
cols = ['g', 'b', 'r']
plt.pie(penguins_group["culmen_length_mm"], labels =
penguins_group['species'],colors = cols)
plt.title('Univariate Analysis of Species', fontsize=15)
plt.show()
plt.savefig("Pie_chart.png")
```
